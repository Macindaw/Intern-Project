**Section 1: Extracting Data From A File**

This section will cover how to read in different file types into your program unsing the Pandas module on python. The Two Types of Files we will learn to read are the 'csv' and 'xlsx' files types. **Expand on Files Types Here**

The two methods we will be using to read in the data are *read_csv("FILE_NAME")* and *read_excel("FILE_NAME)*. These functions will return a data frame upon execution. You can think of a data frame as a sort of excel spreadsheet that contains all of the data in rows and columns. The Rows will each contain a certain type of data and will have a 'title'. This title will be used later to acess certain classes of data.

The Head(n) method return the to 'n' elements of a dataframe. 

```python
#Making the Pandas Module Usable
import pandas as pd

data_csv = pd.read_csv('pokemon_data.csv')
data_exel = pd.read_excel('pokemon_data.xlsx')

print("CSV data")
data_csv.head(10)
```


**Sorting and Manipulating the Data**


Now that you can see that we have sucessfully read in our data and what it looks like it is now time to learn how to manipulate it. The First method we will learn is how to sort the data using *sort_values('title', ascending='')*. The first argument is the title of the row you want to sort by. i.e The bold words ontop of the charts. The other paramater is if you want the data sorted in ascending or decending order and is boolean.

The first task we will do is sort the pokemons by 'HP' from hightest to lowest.

```python
Top_HP = data_csv.sort_values('HP', ascending=False).head(10)
Top_HP
```

If you want to sort data by multiple cells you can use the sort method like this: 

*data_csv.loc[(data_csv['Type 1'] == 'Grass') & (data_csv['Type 2'] == 'Poison')]*

This will return only pokemon that have the data Type of Poison and Grass.
This method can be used with as many conditions are required. The conditions must be seperated by logical operators. Please ask your teacher about them if you require help. 

```python
data_csv.loc[(data_csv['Type 1'] == 'Grass') & (data_csv['Type 2'] == 'Poison')]
```

**Removing Columns that contain empty data**

Here you might notice that for a few pokemon their type 2 in NaN. Now this isn't a pokemon type as far as i am aware so what is the  NaN. NaN is the way pandas and lots of other numerical libraries say that their is no data in this cell. For example, Blissey only has the type of Normal and has no second data type.

Sometimes the rows with NaN variables can cause problems when trying to use the data to form conclusions. Hence pandas has an inbuilt method that remove rows that contain NaN. This can be done with the dropna(axis='rows') method. Note, this method can also be used to remove columns containing NaN by using dropna(axis='columns')

```python
print("Rows not containing any NaN Variables")
Top_HP.dropna(axis='rows')
```

**Combining it all together**

For using the data with most graphing modules, the data must be contained within an array and not a data frame. Lukily the pandas module has a function that will automatically do that for you. *to_numpy()*




Now that you have learnt all of the steps necesary to preparing the data lets combine all the steps necesary to achieve an objective. For this example i will be preparing the data for a graph that will show pokemon type and hp.



```python
poke_data = pd.read_csv('pokemon_data.csv') #Reading in the data from file
extracted = poke_data[['Type 1', 'HP']]  #Extracting the necessary fields
extracted = extracted.dropna(axis='rows')  #Removing unwanted cells
extracted = extracted.sort_values('HP', ascending=False) #Sorting the data by descending HP
graph_data = extracted.to_numpy()
graph_data
```

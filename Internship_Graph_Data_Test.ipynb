**Required Library Imports**


```python
import pandas as pd
import numpy
from matplotlib import pyplot as plt

#Reading in the Data
survey_data = pd.read_csv('surveydata.csv')
survey_data
```

**Creating the First Graphs**

```python
#Creating Data Subset
survey_data_sep1 = survey_data[['How old are you?', 'What time do you try to go to sleep?']].sort_values(['How old are you?', 'What time do you try to go to sleep?'], ascending=[1,0])

#Getting Ride of Columns with NaN Data Type (Data Transforms)
survey_data_sep1 = survey_data_sep1.dropna(axis='columns')
survey_data_sep1 = survey_data_sep1.loc[(survey_data_sep1['How old are you?'] > 5) & (survey_data_sep1['How old are you?'] < 21)]


#Converting to Numpy Data Array
#Getting x and y values for pyplot
dev_x = survey_data_sep1['How old are you?'].to_numpy()
dev_y = survey_data_sep1['What time do you try to go to sleep?'].to_numpy()

plt.plot(dev_x, dev_y)

plt.show()
```
